```python
plt.figure(figsize=(15,5))
plt.subplots_adjust( wspace= 0.5)
plt.subplot(1,4,1)
plt.title ('Actual_MPG')
plt.plot(cars.Actual_MPG,'r')
plt.subplot(1,4,2)
plt.title ('Weight')
plt.plot(cars.Weight, 'b')
plt.subplot(1,4,3)
plt.title ('Cylinders')
plt.plot(cars.Cylinders, 'g')
plt.subplot(1,4,4)
plt.title ('Horsepower')
plt.plot(cars.Horsepower,'y')
plt.savefig('subplot.png')
```

```python
# Heatmap- first load library and datasets
flight_data = sns.load_dataset("flights")
flight_data
#flight_data.shape
```

```python
# rearrange data
flight_data = flight_data.pivot('month','year','passengers')
flight_data.to_csv('flight.csv')
```

```python
# generate map
sns.heatmap(flight_data, annot= True)
plt.show()
```

```python
sns.heatmap(flight_data, cmap="YlGnBu")
plt.show()
```
